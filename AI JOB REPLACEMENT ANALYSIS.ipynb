```python
df["country"] = None
df.head()
df["country"] = ""
df.head()
countries = [
    "USA", "UK", "Canada", "Germany", "Australia",
    "India", "Japan", "Brazil", "Singapore",
    "France", "UAE", "Netherlands", "China",
    "Spain", "Italy"
]
df.country = np.random.choice(countries, size=len(df))
df.head()
country_cities = {

    'USA': ['New York','Los Angeles','Chicago','Houston','San Francisco','Seattle','Boston','Austin'],
    'UK': ['London','Manchester','Birmingham','Leeds','Glasgow','Edinburgh'],
    'Canada': ['Toronto','Vancouver','Montreal','Calgary','Ottawa'],
    'Germany': ['Berlin','Munich','Hamburg','Frankfurt','Cologne'],
    'Australia': ['Sydney','Melbourne','Brisbane','Perth','Adelaide'],
    'India': ['Mumbai','Delhi','Bangalore','Hyderabad','Chennai','Pune','Ahmedabad'],
    'Japan': ['Tokyo','Osaka','Yokohama','Nagoya'],
    'Brazil': ['Sao Paulo','Rio de Janeiro','Brasilia','Salvador'],
    'Singapore': ['Singapore','Jurong East','Tampines'],
    'France': ['Paris','Lyon','Marseille'],
    'UAE': ['Dubai','Abu Dhabi','Sharjah'],
    'Netherlands': ['Amsterdam','Rotterdam','Utrecht'],
    'China': ['Beijing','Shanghai','Shenzhen'],
    'Spain': ['Madrid','Barcelona','Valencia'],
    'Italy': ['Rome','Milan','Naples']
}
df["city"] = df["country"].map(lambda x: random.choice(country_cities[x]))
df.head()
```

```python
experience_levels = ["Entry", "Mid", "Senior", "Lead"]
df["experience_level"] = np.random.choice(experience_levels, size=len(df))
df.head()
```

```python
education_levels = [
    "High School",
    "Bachelor's",
    "Master's",
    "PhD",
    "Bootcamp"
]
df["education_level"] = np.random.choice(education_levels, size=len(df))
df.head()
def generate_experience(row):
    
    level = row["experience_level"]
    edu = row["education_level"]
    
    # Base experience range
    if level == "Entry":
        exp = random.randint(0, 2)
        
    elif level == "Mid":
        exp = random.randint(3, 5)
        
    elif level == "Senior":
        exp = random.randint(6, 9)
        
    elif level == "Lead":
        exp = random.randint(10, 15)
        
    else:
        exp = random.randint(1, 5)
    
    # Education adjustment
    if edu == "Master's":
        exp += 1
    elif edu == "PhD":
        exp += 2
    elif edu == "Bootcamp":
        exp -= 1
    
    # Ensure no negative experience
    exp = max(exp, 0)
    
    return exp
df["years_experience"] = df.apply(generate_experience, axis=1)
df.head()
```

```python
def generate_salary(row):
    level = row["experience_level"]
    exp = row["years_of_experience"]
    
    # Entry Level
    if level == "Entry":
        base = random.randint(30000, 70000)
        
    # Mid Level
    elif level == "Mid":
        base = random.randint(60000, 110000)
        
    # Senior Level
    elif level == "Senior":
        base = random.randint(100000, 180000)
        
    # Lead Level
    elif level == "Lead":
        base = random.randint(150000, 300000)
        
    else:
        base = random.randint(40000, 100000)
    
    # Adjust based on experience
    base += exp * random.randint(1000, 3000)
    
    return base
df["years_of_experience"] = df.apply(generate_experience, axis=1)
df.head()
```

```python
#Salary generation with country adjustment before AI
def base_salary(level):
    
    if level == "Entry":
        return random.randint(30000, 70000)
    
    elif level == "Mid":
        return random.randint(60000, 110000)
    
    elif level == "Senior":
        return random.randint(100000, 180000)
    
    elif level == "Lead":
        return random.randint(150000, 300000)
    
    else:
        return random.randint(40000, 90000)
country_multiplier = {
    "USA": 1.5,
    "UK": 1.3,
    "Germany": 1.2,
    "Canada": 1.3,
    "Australia": 1.3,
    "India": 0.6,
    "Brazil": 0.5,
    "UAE": 1.2,
    "Singapore": 1.4,
    "Japan": 1.2,
    "Spain": 1.1,
    "Italy": 1.1,
    "Netherlands": 1.2,
    "China": 0.9
}
def generate_salary_before(row):
    
    level = row["experience_level"]
    exp = row["years_of_experience"]
    country = row["country"]
    
    # Base salary
    salary = base_salary(level)
    
    # Experience increment
    salary += exp * random.randint(1500, 4000)
    
    # Country adjustment
    salary *= country_multiplier.get(country, 1)
    
    return salary
df["salary_before_usd"] = df.apply(generate_salary_before, axis=1)
df.head()
```

```python
# Salary generation with country adjustment after AI
def generate_salary_after(row):
    
    # AI impact: -10% to +25%
    change_percent = random.uniform(-10, 25)
    
    new_salary = row["salary_before_usd"] * (1 + change_percent/100)
    
    return new_salary, change_percent
results = df.apply(generate_salary_after, axis=1)

df["salary_after_usd"] = [r[0] for r in results]
df["salary_change_percent"] = [r[1] for r in results]
df["salary_before_usd"] = df["salary_before_usd"].round(0)
df["salary_after_usd"] = df["salary_after_usd"].round(0)
df["salary_change_percent"] = df["salary_change_percent"].round(2)
df[[
    "experience_level",
    "years_of_experience",
    "country",
    "salary_before_usd",
    "salary_after_usd",
    "salary_change_percent"
]].head(10)
```

```python
#Data cleaning 
# 1. Remove duplicates
df = df.drop_duplicates()
# 2. Remove rows with missing values
df = df.dropna()
# 2. Fix data types
numeric_cols = df.select_dtypes(include=['float64','int64']).columns
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
# 3. Handle outliers (e.g., salaries above $500,000)
df = df[df["salary_before_usd"] <= 500000]
df[[
    "experience_level",
    "years_of_experience",
    "country",
    "salary_before_usd",
    "salary_after_usd",
    "salary_change_percent"
]].head(10)     
# 3. Standardize text
text_cols = df.select_dtypes(include='object').columns

for col in text_cols:
    df[col] = df[col].str.strip().str.lower()
df[[
    "experience_level",
    "years_of_experience",
    "country",
    "salary_before_usd",
    "salary_after_usd",
    "salary_change_percent"
]].head(10)
# 3. Standardize text
text_cols = df.select_dtypes(include='object').columns

for col in text_cols:
    df[col] = df[col].str.strip().str.lower()
df[[
    "experience_level",
    "years_of_experience",
    "country",
    "salary_before_usd",
        "salary_after_usd", 
    "salary_change_percent"
]].head(10)
df["country"] = df["country"].str.title()
df["experience_level"] = df["experience_level"].str.title()
df["education_level"] = df["education_level"].str.title()
df[[
    "experience_level",
    "years_of_experience",
    "country",
    "salary_before_usd",
    "salary_after_usd",
    "salary_change_percent"
]].head(10)
```

```python
#🚀 PHASE 4 — FINAL FEATURE ENGINEERING (INTERVIEW SAFE)
# Salary impact
df["salary_impact"] = df["salary_after_usd"] - df["salary_before_usd"]

# Salary growth %
df["salary_growth_pct"] = (df["salary_impact"] / df["salary_before_usd"]) * 100

# AI impact score
df["ai_impact_score"] = (
    0.5 * df["automation_risk_percent"] +
    0.3 * df["ai_replacement_score"] +
    0.2 * df["ai_disruption_intensity"]
)

# Skill pressure
df["skill_pressure_index"] = (
    df["skill_gap_index"] + df["reskilling_urgency_score"]
) / 2

# Risk category binary
df["is_high_risk"] = (df["automation_risk_percent"] > 70).astype(int)

# Experience numeric
exp_map = {"Entry":1,"Mid":2,"Senior":3,"Lead":4}
df["experience_numeric"] = df["experience_level"].map(exp_map)

# Salary per experience (VERY IMPORTANT)
df["salary_per_year_exp"] = df["salary_after_usd"] / (df["years_of_experience"] + 1)
```
